```python
def flatten_dict(dd, separator ='_', prefix =''):
    return { prefix + separator + k if prefix else k : v
             for kk, vv in dd.items()
             for k, v in flatten_dict(vv, separator, kk).items()
             } if isinstance(dd, dict) else { prefix : dd }
```

```python
catList = dict()
def catIds(child):
    global catList
    if isinstance(child,dict):
        if child["name"] in catList.keys():
            if child['product_count']>0:
                child["name"] = child["id"]
        else:
            catList[child["name"]] = child["id"]
        if len(child["children_data"])==0:
            return
        else:
            catIds(child["children_data"])
    else:
        for i in child:
            catIds(i)

catIds(jsonResp)
```
