```python
import pandas as pd
import numpy as np
import re
```

```python
db = pd.read_excel("Data/export_Product_Xlsx_Export_2023-02-14_04-56-50_2.xlsx")
```

```python
df = pd.DataFrame(db,index=None)
df.head()
```

```python
new_length = []
new_width = []
for i in range(len(df)):
    row = df.iloc[i,:]
    element = row["length"]
    width = row["width"]
    if not pd.isna(element):
        new_element = element + "in"
        new_length.append(new_element)
    else:
        new_length.append(element)
    if not pd.isna(width):
        new_w = width + "in"
        print(new_w)
        new_width.append(new_w)
    else:
        new_width.append(width)
```

```python
df['width'] = new_width
```

```python
new_thickness = []
for i in range(len(df)):
    row = df.iloc[i,:]
    element = row["thickness"]
    if pd.isna(element):
        new_thickness.append(element)
    else:
        if re.match("\d+(_\d)*$",element):
            new_element = element + "in"
            new_element.replace("_0_in","in")
            new_thickness.append(new_element)
        elif re.match("\d+_\w+$",element):
            new_element = element.lower()
            new_thickness.append(new_element)
        else:
            print(element)
        
```

```python
df['thickness'] = new_thickness
```

```python
from fractions import Fraction

def dec_to_proper_frac(dec):
    sign = "-" if dec < 0 else ""
    frac = Fraction(str(abs(dec)))
    if dec<1:
        return (f"{frac.numerator % frac.denominator}/{frac.denominator}")
    return (f"{sign}{frac.numerator // frac.denominator} "
            f"{frac.numerator % frac.denominator}/{frac.denominator}")
```

```python
uni_length = list(df['length'].unique())
uni_length.remove(np.nan)
export_length = pd.DataFrame([],columns=["code","label-en_US","attribute","sort_order"])

codes = []
labels = []
for i in uni_length:
    codes.append(i)
    dec = re.sub("[a-z]+$","",i)
    dec = dec.replace("_",".")
    if float(dec)%1!=0:
        labels.append(str(dec_to_proper_frac(float(dec)))+"\"")
    else:
        labels.append(str(int(dec))+"\"")

export_length["code"] = codes
export_length["label-en_US"] = labels
export_length["attribute"] = "length"
```
